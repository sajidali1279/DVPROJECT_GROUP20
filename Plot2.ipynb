```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data=pd.read_csv('plotvis2.csv')
data.head()
```

```python
import matplotlib.pyplot as plt

# Filter out rows with NaN values in the 'life_expectancy' column
data_filtered = data.dropna(subset=['life_expectancy'])

# Convert 'continent' column to string type
data_filtered['continent'] = data_filtered['continent'].astype(str)

# Bar chart
plt.figure(figsize=(10, 6))
plt.bar(data_filtered['continent'], data_filtered['life_expectancy'], color='brown')
plt.title('Bar Chart of Life Expectancy by Continent')
plt.xlabel('Continent')
plt.ylabel('Life Expectancy')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt
import mpld3

# Filter out rows with NaN values in the 'life_expectancy' column
data_filtered = data.dropna(subset=['life_expectancy'])

# Convert 'continent' column to string type
data_filtered['continent'] = data_filtered['continent'].astype(str)

# Bar chart
plt.figure(figsize=(10, 6))
plt.bar(data_filtered['continent'], data_filtered['life_expectancy'], color='brown')
plt.title('Bar Chart of Life Expectancy by Continent')
plt.xlabel('Continent')
plt.ylabel('Life Expectancy')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()

# Save the plot as an image file
plt.savefig('vis1.png')

# Convert matplotlib plot to HTML using mpld3
html_content = mpld3.fig_to_html(plt.gcf())

# Save the HTML content to a file
with open('vis1.html', 'w') as f:
    f.write(html_content)

# Show the plot
plt.show()
```
