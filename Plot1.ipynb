```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (assuming it's stored in a CSV file)
# Replace 'path_to_your_file.csv' with the actual path to your dataset file
url= 'https://raw.githubusercontent.com/lakshmi-keerthi/lakshmi-keerthi.github.io/main/covid_data_cleaned.csv'
data = pd.read_csv(url)

data
```

```python
data.info()
```

```python
data_new = data[['continent', 'total_cases']].copy()
data_new.columns = ['continent', 'total_cases']  # Rename the columns as per your preference

# Display the first few rows of the new dataframe
print(data_new.head())
```

```python
import matplotlib.pyplot as plt

# Aggregate total cases by continent
total_cases_by_continent = data_new.groupby('continent')['total_cases'].sum()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(total_cases_by_continent, labels=total_cases_by_continent.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Cases Distribution by Continent')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
